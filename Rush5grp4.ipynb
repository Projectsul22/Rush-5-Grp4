```python
import pandas as pd

# Charger le fichier CSV de la multinationale
file_path = 'C:/Users/sybah/Documents/Ecole/Rush5/Camp_Market.csv'
data = pd.read_csv(file_path)

# Afficher les premières lignes pour voir la structure des données
data.head()
```

```python
# Rajout du séparateur ','
data = pd.read_csv(file_path, sep=';')

# Afficher les premières lignes pour vérifier la correction
data.head()
```

```python

# Vérification des valeurs aberrantes dans les colonnes Year_Birth et Income

# Statistiques descriptives pour Year_Birth et Income
year_birth_stats = data['Year_Birth'].describe()
income_stats = data['Income'].describe()

# Vérifier s'il y a des âges incohérents (par exemple, trop jeunes ou trop vieux)
# Calculer l'âge en fonction de l'année de naissance
current_year = 2024
data['Age'] = current_year - data['Year_Birth']

# Statistiques descriptives sur l'âge
age_stats = data['Age'].describe()

# Afficher les résultats des statistiques descriptives
year_birth_stats, income_stats, age_stats
```

```python
# Identifier les valeurs aberrantes dans la colonne Age (âges supérieurs à 100 ans)
outliers_age = data[data['Age'] > 100]

# Identifier les valeurs aberrantes dans la colonne Income (revenu trop bas ou trop élevé)
outliers_income = data[(data['Income'] < 2000) | (data['Income'] > 300000)]

# Supprimer les valeurs aberrantes sur l'âge (supérieur à 100 ans)
data_cleaned = data[data['Age'] <= 100]

# Supprimer les valeurs aberrantes sur le revenu (moins de 2000 et plus de 300000)
data_cleaned = data_cleaned[(data_cleaned['Income'] >= 2000) & (data_cleaned['Income'] <= 300000)]

# Indiquer les valeurs supprimées
outliers_age_list = outliers_age[['ID', 'Year_Birth', 'Age']]
outliers_income_list = outliers_income[['ID', 'Income']]

# Afficher les valeurs aberrantes supprimées pour référence
outliers_age_list, outliers_income_list
```
