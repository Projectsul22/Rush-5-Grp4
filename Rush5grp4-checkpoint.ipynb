```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# Rajout du séparateur ','
file_path='Camp_Market.csv'
data = pd.read_csv(file_path, delimiter=';')

# Afficher les premières lignes pour vérifier la correction
data.head()
```

```python

# Vérification des valeurs aberrantes dans les colonnes Year_Birth et Income

# Statistiques descriptives pour Year_Birth et Income
year_birth_stats = data['Year_Birth'].describe()
income_stats = data['Income'].describe()

# Vérifier s'il y a des âges incohérents (par exemple, trop jeunes ou trop vieux)
# Calculer l'âge en fonction de l'année de naissance
current_year = 2024
data['Age'] = current_year - data['Year_Birth']

# Statistiques descriptives sur l'âge
age_stats = data['Age'].describe()

# Afficher les résultats des statistiques descriptives
year_birth_stats, income_stats, age_stats
```

INTERPRETATION :
--------------
Âge moyen : 55 ans, avec une diversité d'âges (écart-type de 11.9 ans).
Revenu moyen : 52 247 €, avec une grande variation (écart-type de 25 517 €).
Valeurs possiblement aberrantes : Des revenus extrêmes (min 1 730 €, max 666 666 €) et des âges potentiellement incohérents doivent être vérifiés.

```python
# Identifier les valeurs aberrantes dans la colonne Age (âges supérieurs à 100 ans)
outliers_age = data[data['Age'] > 100]

# Identifier les valeurs aberrantes dans la colonne Income (revenu trop bas ou trop élevé)
outliers_income = data[(data['Income'] < 2000) | (data['Income'] > 300000)]

# Supprimer les valeurs aberrantes sur l'âge (supérieur à 100 ans)
data_cleaned = data[data['Age'] <= 100]

# Supprimer les valeurs aberrantes sur le revenu (moins de 2000 et plus de 300000)
data_cleaned = data_cleaned[(data_cleaned['Income'] >= 2000) & (data_cleaned['Income'] <= 300000)]

# Indiquer les valeurs supprimées
outliers_age_list = outliers_age[['ID', 'Year_Birth', 'Age']]
outliers_income_list = outliers_income[['ID', 'Income']]

# Afficher les valeurs aberrantes supprimées pour référence
outliers_age_list, outliers_income_list
```

INTERPRETATION :
--------------
Valeurs aberrantes sur l'âge : Les clients de plus de 100 ans sont identifiés comme aberrants (ex : client de 123 ans).
Valeurs aberrantes sur le revenu : Les revenus inférieurs à 2 000 € ou supérieurs à 300 000 € sont supprimés (ex : revenus de 1 730 € et 666 666 €).
Nettoyage des données : Les lignes avec ces valeurs extrêmes ont été supprimées pour améliorer la qualité des analyses futures.

```python
# ANALYSE TAUX DE CONVERSION DES CAMPAGNES MARKETING
```

Le ROI (Return on Investment) des campagnes marketing est un indicateur clé qui permet de mesurer l'efficacité d'une campagne en termes de retour financier par rapport aux coûts engagés.

```python
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
total_clients = len(data_cleaned)

# Calcul du pourcentage de clients ayant accepté chaque campagne
conversion_rates = data_cleaned[campaign_columns].mean() * 100

# Afficher les taux de conversion pour chaque campagne
print(conversion_rates)

# Tracer les résultats pour visualiser les taux de conversion
conversion_rates.plot(kind='bar', figsize=(10, 6), edgecolor='black')
plt.title('Taux de conversion des campagnes marketing')
plt.xlabel('Campagnes')
plt.ylabel('Taux de conversion (%)')
plt.xticks(rotation=45)
plt.show()
```

INTERPRETATION : 
--
Taux de conversion des différentes campagnes marketing. Voici les points clés :

    Campagnes avec le meilleur taux de conversion :
        La campagne globale, intitulée "Response", a le taux de conversion le plus élevé avec environ 15% des clients ayant accepté au moins une campagne. Cela montre une efficacité relativement bonne lorsque l'on considère l'ensemble des campagnes.
        Les campagnes spécifiques AcceptedCmp3, AcceptedCmp4, et AcceptedCmp5 ont des taux de conversion autour de 7%, suggérant qu'elles ont eu un impact modéré.

    Campagnes avec des taux de conversion plus faibles :
        La campagne AcceptedCmp2 a le plus faible taux de conversion, avec seulement environ 1,3% de participation, indiquant qu'elle a eu très peu d'impact sur les clients.

Conclusion : La campagne générale "Response" a un impact global positif, mais il existe une disparité notable dans l'efficacité des différentes campagnes spécifiques, avec AcceptedCmp2 nécessitant potentiellement une révision ou un repositionnement.

```python
# Liste des campagnes
campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']

# Initialiser un dictionnaire pour stocker les résultats du ROI
roi_per_campaign = {}

# Coût par contact (constant selon les données)
cost_per_contact = data_cleaned['Z_CostContact'].unique()[0]  # On suppose que c'est constant

# Calculer le ROI pour chaque campagne
for campaign in campaigns:
    # Nombre de participants pour chaque campagne
    num_participants = data_cleaned[campaign].sum()
    
    # Coût total de la campagne
    total_cost = num_participants * cost_per_contact
    
    # Gains totaux (dépenses des participants dans les différentes catégories de produits)
    gains = data_cleaned[data_cleaned[campaign] == 1][['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum().sum()
    
    # Calcul du ROI
    if total_cost > 0:  # Éviter une division par zéro
        roi = ((gains - total_cost) / total_cost) * 100
    else:
        roi = 0
    
    # Ajouter le ROI au dictionnaire
    roi_per_campaign[campaign] = roi

# Afficher les résultats du ROI pour chaque campagne
roi_per_campaign
```

Coût total : Calculé en multipliant le nombre de participants par le coût par contact (Z_CostContact).

Gains totaux : Somme des dépenses des clients ayant participé à chaque campagne dans les différentes catégories de produits.

ROI extrêmement élevé :

    Toutes les campagnes montrent des ROI très élevés, bien supérieurs à 100%. Cela signifie que les gains générés par les clients ayant participé aux campagnes dépassent largement le coût des campagnes.
    Le ROI le plus élevé est observé pour AcceptedCmp5 (53,732%), ce qui indique que cette campagne a été la plus rentable parmi toutes celles analysées.

Comparaison des campagnes :

    AcceptedCmp1, AcceptedCmp5, et AcceptedCmp2 se distinguent avec des ROI autour de 50%, ce qui montre qu'elles ont été particulièrement efficaces en termes de retour financier.
    AcceptedCmp3 a le ROI le plus bas (23,918%), ce qui pourrait indiquer que, bien qu'elle ait généré des gains, elle a été moins efficace par rapport aux autres campagnes.

Réponse à la dernière campagne :

    Le ROI de la dernière campagne (32,755%) est également positif, ce qui montre que cette campagne a généré des retours financiers significatifs.

Explication du ROI :

    Un ROI de 100% signifie que les gains de la campagne sont exactement égaux à son coût (les gains couvrent les dépenses, mais rien de plus).

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# 1. Préparation des données
# Sélectionner les caractéristiques pertinentes
features = ['Age', 'Income', 'Kidhome', 'Teenhome', 'MntWines', 'MntFruits', 
            'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
X = data_cleaned[features]  # Caractéristiques (X)
y = data_cleaned['AcceptedAnyCampaign'].astype(int)  # Variable cible (réponse à une campagne)

# Diviser les données en jeu d'entraînement et jeu de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Entraînement du modèle Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 3. Prédiction sur le jeu de test
y_pred = rf_model.predict(X_test)

# 4. Évaluation des performances
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1])

# Affichage des résultats
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"AUC-ROC: {roc_auc:.2f}")

# 5. Importance des variables
importances = pd.Series(rf_model.feature_importances_, index=features)
importances.sort_values().plot(kind='barh', figsize=(10,6), title='Importance des caractéristiques')
```

```python
from sklearn.metrics import confusion_matrix

# Prédiction sur le jeu de test
y_pred_rf = rf_model.predict(X_test)

# Calcul de la matrice de confusion
cm = confusion_matrix(y_test, y_pred_rf)

# Visualisation de la matrice de confusion
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Non réactif", "Réactif"], 
            yticklabels=["Non réactif", "Réactif"])
plt.title("Matrice de confusion - Random Forest (Prédiction réponse campagne)")
plt.ylabel("Valeurs réelles")
plt.xlabel("Valeurs prédites")
plt.show()
```

```python
# Importation de SMOTE pour le rééquilibrage
from imblearn.over_sampling import SMOTE

# Rééquilibrage des données avec SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Entraînement d'un nouveau modèle Random Forest sur les données rééquilibrées
rf_model_resampled = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_resampled.fit(X_resampled, y_resampled)

# Prédiction sur le jeu de test
y_pred_resampled = rf_model_resampled.predict(X_test)

# Évaluation des performances après rééquilibrage
accuracy_resampled = accuracy_score(y_test, y_pred_resampled)
precision_resampled = precision_score(y_test, y_pred_resampled)
recall_resampled = recall_score(y_test, y_pred_resampled)
f1_resampled = f1_score(y_test, y_pred_resampled)
roc_auc_resampled = roc_auc_score(y_test, rf_model_resampled.predict_proba(X_test)[:, 1])

# Affichage des nouvelles métriques
print(f"Accuracy après rééquilibrage: {accuracy_resampled:.2f}")
print(f"Precision après rééquilibrage: {precision_resampled:.2f}")
print(f"Recall après rééquilibrage: {recall_resampled:.2f}")
print(f"F1-Score après rééquilibrage: {f1_resampled:.2f}")
print(f"AUC-ROC après rééquilibrage: {roc_auc_resampled:.2f}")


# Création de la matrice de confusion
cm_resampled = confusion_matrix(y_test, y_pred_resampled)

# Visualisation de la matrice de confusion avec un heatmap
plt.figure(figsize=(8,6))
sns.heatmap(cm_resampled, annot=True, fmt="d", cmap="Blues", xticklabels=["Non réactif", "Réactif"], 
            yticklabels=["Non réactif", "Réactif"])
plt.title("Matrice de confusion après rééquilibrage - Random Forest")
plt.ylabel("Valeurs réelles")
plt.xlabel("Valeurs prédites")
plt.show()
```
